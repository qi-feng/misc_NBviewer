```python
%matplotlib inline
#!/usr/bin/python
import math
import numpy
import matplotlib.pyplot as plt

#SI unit
#Planck's constant unit m2 kg / s
#Boltzmann constant unit m2 kg s-2 K-1
h=6.62606957e-34
c=299792458.0
k=1.3806488e-23

#Planck's law in terms of wavelength
#specific intensity B_lambda 
#SI unit: W sr-1 m-3 

def PlanckFunc(mylambda,T):
  B_lambda=2.0*h*c*c/((mylambda**5)*(math.exp(h*c/(mylambda*k*T))-1.0))
  return B_lambda

def GetIndex400_600nm(T):
  B600=PlanckFunc(600.e-9,T)
  B400=PlanckFunc(400.e-9,T)
  Index=(B600-B400)/200.e-9
  #Index=(600e-9*B600-400e-9*B400)/200e-9
  return Index

def GetC400_600nm(T):
  logB600=math.log10(PlanckFunc(600.e-9,T))
  logB400=math.log10(PlanckFunc(400.e-9,T))
  #C=(logB600-logB400)/200.e-9
  C=(logB600-logB400)
  return C

#fig = plt.figure(figsize=(6, 5))
#fig = plt.figure()
#ax1 = fig.add_subplot(1,1,1)

#test=PlanckFunc(0.4e-6,7000)
#print '0.4 um 7000K test'
#print test

#Create an array of temperatures
tlist=numpy.arange(4000, 8000, 1000)
#tlist=numpy.arange(4000, 8000, 1000)

indexlist=[]
Blist=[]
clist=[]

# create a list of wavelength in unit of nm for plotting
wavelist=numpy.arange(20e-9,1600e-9,20e-9)

for temp in tlist:
    Blist=[]
    for wave in wavelist:
        Blist.append(PlanckFunc(wave,temp))
        #Blist.append(math.log10(PlanckFunc(wave,7000)))

    plt.plot(wavelist*1e9,Blist)
    #plt.semilogy(wavelist*1e6,Blist)
    plt.annotate("%s K" % temp, xy=(wavelist[40]*1e9,Blist[40]),xycoords='data', xytext=(wavelist[40]*1e9*1.2,Blist[40]*1.8),arrowprops=dict(arrowstyle="->", facecolor='black'))

#print indexlist

#plt.annotate("%s" % temp, xy=(0.5,0.5),xycoords='axes fraction', xytext=(0.6,0.6),arrowprops=dict(arrowstyle="->", facecolor='black'))
#plt.annotate("%s" % temp, xy=(wavelist[40]*1e6,Blist[40]),xycoords='data', xytext=(wavelist[40]*1e6*1.2,Blist[40]*1.2),arrowprops=dict(arrowstyle="->", facecolor='black'))

plt.plot(wavelist*1e9,Blist)
#plt.semilogy(tlist,indexlist)
#plt.plot(tlist,clist)
plt.ylabel('specific intensity (W sr$^{-1}$ m$^{-3}$)')
plt.xlabel('wavelength (nm)')
plt.show()

#plt.savefig('PlanckBlackBody.eps',format='eps', dpi=1000)



```
